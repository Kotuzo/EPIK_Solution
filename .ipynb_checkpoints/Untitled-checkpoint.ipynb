```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report,confusion_matrix
import os
```

```python
mapper_target_col = {
    'f': 0,
    't': 1
}
```

```python
remove_cols = ['business', 'private', 'arranged', 'exchange', 'free', 'price', 'derivative']
def load_train_data(dropna=False):
    test_filename = 'transformed2017_10.csv'
    october = 'transformed2016_11.csv'
    predict_cols = ['predict_sold', 'predict_replies', 'predict_views']
    data = []
    for file in os.listdir(DATA_PATH):
        if file != test_filename or file != october:
            data.append(pd.read_csv(DATA_PATH + file))
    df = pd.concat(data)
    if dropna:
        df = df.dropna()
    df_X = df.drop(['id', 'used'] + predict_cols + remove_cols, 1)
    df_Y = df['predict_sold']
    df_Y = df_Y.map(mapper_target_col)
    return df_X, df_Y

def load_test_data():
    test_filename = 'transformed2017_10.csv'
    df = pd.read_csv(DATA_PATH + 'transformed2017_10.csv')
    return df.drop(['used'] + remove_cols, 1)
```

```python
X, Y = load_train_data(True)
```

```python
X_tr, X_test, Y_tr, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
print(X_tr.head())
print(len(X_tr))
```

```python
X_Y_tr =  X_tr
X_Y_tr['predict_sold'] = Y_tr
```

```python
clf = RandomForestClassifier(n_estimators=128, verbose=True, n_jobs=3)
```

```python
X_tr.head()
```

```python
Y_tr.head()
```

```python
clf.fit(X_tr, Y_tr)
```

```python
preds = clf.predict(X_test)
```

```python
print(classification_report(Y_test, preds))
```

```python
print(confusion_matrix(Y_test,preds))
```

```python
accuracies = cross_val_score(estimator=clf, X=X_tr, y=Y_tr, cv=7, n_jobs=-1)
```

```python
print(accuracies)
accuracies.mean()
```

```python
X_T = load_test_data()
```

```python
X_T = X_T.fillna(0)
```

```python
X_T.head()
```

```python
predictions = clf.predict(X_T.drop(['id'], 1))
```

```python
sum(predictions)
```

```python
output = ''
for ajdik, p in zip(X_T['id'], predictions):
    output += str(ajdik) + ',' + str(p) + ',0,0\n'
```

```python
with open('sub1.csv', 'w') as F:
    F.write(output)
```
